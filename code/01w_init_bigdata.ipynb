# 인구 데이터 전처리
- 첫 열은 index로, 콤마가 들어간 값은 정수로 파일을 읽어 pop에 저장
- 키워드 인자 encoding = 'cp949', index_col=0, thousands=','를 사용

```python
import pandas as pd
pop = pd.read_csv("popmonth.csv", encoding='cp949', index_col=0, thousands=',')
pop
```

```python
# 데이터프레임 pop의 열이름을 수정
# 첫 5행을 표시

pop.columns = ['총인구수', '세대수','세대당인구','남자인구수','여자인구수','남여비율']
pop.head()
```

```python
pop.columns[2]
```

```python
pop.index
```

```python
[pop.index[i] for i in range(len(pop))]
```

```python
# pop의 index를 수정 - 행정구역 코드 제거

#도시이름만으로 구성

pop.index = [pop.index[i].split()[0] for i in range(len(pop))]
pop
```

```python
# 인구 데이터 시각화
# seaborn의 sns.scatterplot()으로 '총인구수'의 산점도를 그린코드

import matplotlib.pyplot as plt
import seaborn as sns
sns.scatterplot([pop['총인구수']])
plt.xticks(rotation=90)
plt.show()
```

```python
pop.sort_values('총인구수') # '총인구수' 로 정렬
```

```python
# '총인구수'로 정렬해 막대 그래프를 그린 코드

import matplotlib.pyplot as plt
import seaborn as sns
pop1 = pop.sort_values('총인구수')
sns.barplot(x=pop1.index, y='총인구수', data=pop1)
plt.xticks(rotation=90)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(5, 3)) # 사이즈 조절


pop1 = pop.sort_values('총인구수')
sns.barplot(x=pop1.index, y='총인구수', data=pop1)
plt.xticks(rotation=90)
plt.show()
```
