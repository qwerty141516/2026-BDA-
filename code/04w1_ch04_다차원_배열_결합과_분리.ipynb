# np.column_stack( ( a, b ) )

> 2차원 배열 a,b를 수평으로 열(coulmns wise)로 결합

```python
x = np.array([ 1, 2, 3 ])
y = np.array([ 10, 20, 30 ])
np.column_stack((x, y))

#np.column_stack((x[:, np.newaxis], y[:, np.newaxis])) 와 결과가 동일
```

```python
x[:, np.newaxis]

# (3, ) 에서 (3, 1)로 변환
```

```python
np.hstack((x, y))

# 일반적으로 hstack()는 column_stack와 다름
```

# np.c_[]
> 함수는 아니고 대괄호 인덱스 []를 사용해 배열 객체 반환

- 대괄호 인덱싱 문법을 통해 열 방향으로 배열을 결합하는 도구

- 배열을 열 방향(2번째 축)으로(concatenation along the second axis, 즉 column-wise) 연결
  - 슬라이싱 문법을 지원하여, 인덱싱이나 슬라이싱과 함께 사용 가능

- 주요 특징
  - 배열 결합
    - 여러 배열을 수평(가로) 방향으로 쉽게 결합
  - 1차원 배열 확장
    - 1차원 배열을 자동으로 2차원으로 확장한 뒤 열로 결합

```python
import numpy as np

# 두 개의 1D 배열을 열 방향으로 결합

a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

result = np.c_[a, b]
print(result)

# 이 코드는 두 1차원 배열 [1, 2, 3]과 [4, 5, 6]을
# 세로로 결합하여 다음과 같은 결과를 출력합니다
```

# 난수 제너레이터의 함수 integers()

- rg.intergers(low=0, high=10, size=3)
  - [low, high)에서 난수 3개
    - low, low + 1, low + 2, ... < high 시퀀스 정수 중에서 난수 3개
    

```python
import numpy as np

rg = np.random.default_rng(12345)
r = rg.integers(low=0, high=10, size=3)
r
```

```python
rg.integers(1, 7, (3, 4))
```

# 난수 생성자 함수 standard_normal()

> 정규 분포 난수 발생
  > 표준 정규분포: 68, 95, 99.7

- rg.standard_normal(10)
  - 정규 분포 난수 10개

```python
import numpy as np

rg = np.random.default_rng(seed=42)
rg.standard_normal(10)
```

```python
rg.standard_normal((4, 3))
```
